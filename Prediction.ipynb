```python
import csv
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt 
```

```python
train_features = pd.read_csv("q2-train-features.csv") 
train_features = train_features.iloc[:,1:5].values
train_features = train_features.astype('float64')
train_labels = train_features[:,1:2]
```

```python
len(train_features)
```

```python
#Question 2 Part B

train_features_first = train_features[:,0:1]
temp_arr = np.ones((len(train_features_first),1))
arr_one = np.concatenate((temp_arr, train_features_first), axis = 1)
```

```python
#Question 2 Part B Finding the weights

transposeX = np.transpose(arr_one)
first_mult = np.dot(transposeX, arr_one)
inverse = np.linalg.inv(first_mult)
w_temp = np.dot(inverse, transposeX)
w0 = np.dot(w_temp, train_labels)
print("W0 and W1: " + str(w0))
```

```python
result = np.dot( arr_one, w0)
plt.plot(train_features[:,0:1], result, label = "Predictions")
plt.plot(train_features[:,0:1], train_features[:,1:2], label = "Real Data") 
plt.xlabel('Months') 
plt.ylabel('USD') 
plt.title('Prediction vs Real Data') 
MSEVAL = np.dot(np.transpose(train_labels-result),(train_labels-result)) / len(result)
print("mse: " + str(MSEVAL))
```
